```python
import pandas as pd
```

```python
wikipedia_datset = pd.read_csv('wikipedia/wikipedia_dataset.csv')
quora_dataset = pd.read_csv('quora/filtered_quora.csv')
abstracts_dataset = pd.read_csv('abstracts/filtered_abstracts.csv')
gpt_dataset = pd.read_csv('gpt/filtered_gpt_responses.csv')
claude_dataset = pd.read_csv('claude/filtered_claude_responses.csv')
deepseek_dataset = pd.read_csv('deepseek/filtered_train.csv')
```

```python
wikipedia_datset.head()
wikipedia_datset.rename(columns={'text': 'Text'}, inplace=True)
wikipedia_datset['Label'] = 'Human'
wikipedia_datset['Source'] = 'Wikipedia'
wikipedia_datset = wikipedia_datset[['Text', 'Label', 'Source']]
wikipedia_datset.head()
```

```python
quora_dataset.head()
quora_dataset.rename(columns={'answer': 'Text'}, inplace=True)
quora_dataset['Label'] = 'Human'
quora_dataset['Source'] = 'Quora'
quora_dataset = quora_dataset[['Text', 'Label', 'Source']]
quora_dataset.head()
```

```python
abstracts_dataset.head()
abstracts_dataset.rename(columns={'abstract': 'Text'}, inplace=True)
abstracts_dataset['Label'] = 'Human'
abstracts_dataset['Source'] = 'Abstract'
abstracts_dataset = abstracts_dataset[['Text', 'Label', 'Source']]
abstracts_dataset.head()
```

```python
gpt_dataset.head()
gpt_dataset.rename(columns={'abstract': 'Text'}, inplace=True)
gpt_dataset['Label'] = 'AI'
gpt_dataset['Source'] = 'GPT'
gpt_dataset = gpt_dataset[['Text', 'Label', 'Source']]
gpt_dataset.head()
```

```python
claude_dataset.head()
claude_dataset.rename(columns={'abstract': 'Text'}, inplace=True)
claude_dataset['Label'] = 'AI'
claude_dataset['Source'] = 'Claude'
claude_dataset = claude_dataset[['Text', 'Label', 'Source']]
claude_dataset.head()
```

```python
deepseek_dataset.head()
deepseek_dataset.rename(columns={'content': 'Text'}, inplace=True)
deepseek_dataset['Label'] = 'AI'
deepseek_dataset['Source'] = 'DeepSeek'
deepseek_dataset = deepseek_dataset[['Text', 'Label', 'Source']]
deepseek_dataset.head()
```

```python
# Join all datasets
all_datasets = pd.concat([wikipedia_datset, quora_dataset, abstracts_dataset, gpt_dataset, claude_dataset, deepseek_dataset])
all_datasets.sample(10)
```

```python
# Picking 1000 samples from each source
sample_size = 1000
sampled_datasets = all_datasets.groupby('Source').apply(lambda x: x.sample(sample_size)).reset_index(drop=True)
sampled_datasets.sample(10)
```

```python
sampled_datasets.shape
sampled_datasets = sampled_datasets[['Text', 'Label']]

sampled_datasets.head()

```

```python
# Clean non-ascii characters
sampled_datasets['Text'] = sampled_datasets['Text'].apply(lambda x: x.encode('ascii', 'ignore').decode())
```

```python
# Encode 0 for Human and 1 for AI
sampled_datasets['Label'] = sampled_datasets['Label'].apply(lambda x: 0 if x == 'Human' else 1)
```

```python
# Check label distribution
sampled_datasets['Label'].value_counts()
```

```python
# Cleaning text
import re
def clean_text(text):
    text = re.sub(r'[^\w\s]', '', text, re.UNICODE)
    text = text.lower()
    return text

sampled_datasets['Text'] = sampled_datasets['Text'].apply(lambda x: clean_text(x))
```

```python
# Check label distribution
sampled_datasets['Label'].value_counts()
```

```python
sampled_datasets.to_csv('sampled_datasets.csv', index=False, sep=',')
```
